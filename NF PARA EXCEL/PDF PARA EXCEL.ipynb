```python
import pdfminer.high_level as pdf
import pandas as pd
import xml.etree.ElementTree as ET

def Extrair_Modelo_SP(pdf_file, output_file):
    # Extrai o texto do PDF
    text = pdf.extract_text(pdf_file)

    # Cria um dicionário para armazenar as informações extraídas
    data = {
        'Nome do Tomador': [],
        'CNPJ do Tomador': [],
        'Nome do Prestador': [],
        'CNPJ do Prestador': [],
        'Número da Nota': [],
        'Data de Emissão': [],
        'Valor Total': [],
        'Código de Serviço': [],
        'Alíquota ISS': [],
        'Valor ISS': [],
        'Valor IRRF': [],
        'Valor CSLL': [],
        'Valor PIS': [],
        'Valor COFINS': [],
        'Valor INSS': [],
        #'Valor ISS retido': [],
    }

    # Extrai as informações do texto
    lines = text.split('\n')
    for i in range(len(lines)):
        
        if lines[i].startswith('TOMADOR DE SERVIÇOS'):
            data['Nome do Tomador'].append(' '.join(lines[i+2].split()[2:5]))
            data['CNPJ do Tomador'].append(' '.join(lines[i+3].split()[1]))
            
        elif lines[i].startswith('PRESTADOR DE SERVIÇOS'):
            data['Nome do Prestador'].append(' '.join(lines[19].split()[1]))
            data['CNPJ do Prestador'].append(' '.join(lines[20].split()[1:5]))

        elif lines[i].startswith('Número da Nota'):
            data['Número da Nota'].append(lines[i+1].split()[-1])

        elif lines[i].startswith('Data e Hora de Emissão'):
            data['Data de Emissão'].append(lines[i+1].split()[0])

        elif lines[i].startswith('VALOR TOTAL DA NOTA'):
            data['Valor Total'].append(lines[i].split()[-1])
            
        elif lines[i].startswith('Código do Serviço'):
            data['Código de Serviço'].append(' '.join(lines[i+1].split()[2:15]))

        elif lines[i].startswith('Alíquota'):
            data['Alíquota ISS'].append(lines[89].split()[0])

        elif lines[i].startswith('Valor do ISS'):
            data['Valor ISS'].append(lines[83].split()[0])
        
        elif lines[i].startswith('INSS'):
            data['Valor INSS'].append(lines[i+4].split()[0])
        
        elif lines[i].startswith('IRRF'):
            data['Valor IRRF'].append(lines[i+10].split()[0])
        
        elif lines[i].startswith('CSLL'):
            data['Valor CSLL'].append(lines[i+10].split()[0])
        
        elif lines[i].startswith('COFINS'):
            data['Valor COFINS'].append(lines[i+10].split()[0])
            
        elif lines[i].startswith('PIS'):
            data['Valor PIS'].append(lines[i+8].split()[0])
        
        
    # Cria o DataFrame
    df = pd.DataFrame(data)

    # Salva os dados no arquivo Excel
    df.to_excel(output_file, encoding='utf-8', index=False)

    print(f"Arquivo '{output_file}' criado com sucesso!")
Extrair_Modelo_SP('NFS_001283_THINK DIGITAL_2.307,68.pdf', 'dados_extraidos.xlsx')
```

```python
import pdfminer.high_level as pdf
import pandas as pd
import re

def Extrair_Modelo_SP(pdf_file, output_file):
    # Extrai o texto do PDF
    text = pdf.extract_text(pdf_file)

    # Cria uma lista para armazenar os dados extraídos
    extracted_data = []

    # Extrai as informações do texto usando expressões regulares
    tomador_regex = r'TOMADOR DE SERVIÇOS\n(.+)\nCNPJ: (\d+)'
    tomador_match = re.search(tomador_regex, text)
    if tomador_match:
        extracted_data.append(('Nome do Tomador', tomador_match.group(1)))
        extracted_data.append(('CNPJ do Tomador', tomador_match.group(2)))
        
    prestador_regex = r'PRESTADOR DE SERVIÇOS\n(.+)\nCNPJ: (\d+)'
    prestador_match = re.search(prestador_regex, text)
    if prestador_match:
        extracted_data.append(('Nome do Prestador', prestador_match.group(1)))
        extracted_data.append(('CNPJ do Prestador', prestador_match.group(2)))
        
    nota_regex = r'Número da Nota\n(\d+)'
    nota_match = re.search(nota_regex, text)
    if nota_match:
        extracted_data.append(('Número da Nota', nota_match.group(1)))
        
    emissao_regex = r'Data e Hora de Emissão\n(.+)'
    emissao_match = re.search(emissao_regex, text)
    if emissao_match:
        extracted_data.append(('Data de Emissão', emissao_match.group(1)))
        
    total_regex = r'VALOR TOTAL DA NOTA\n([\d,.]+)'
    total_match = re.search(total_regex, text)
    if total_match:
        extracted_data.append(('Valor Total', total_match.group(1)))
        
    servico_regex = r'Código do Serviço\n(.+)'
    servico_match = re.search(servico_regex, text)
    if servico_match:
        extracted_data.append(('Código de Serviço', servico_match.group(1)))
        
    aliquota_regex = r'Alíquota\n([\d,.]+)'
    aliquota_match = re.search(aliquota_regex, text)
    if aliquota_match:
        extracted_data.append(('Alíquota ISS', aliquota_match.group(1)))
        
    iss_regex = r'Valor do ISS\n([\d,.]+)'
    iss_match = re.search(iss_regex, text)
    if iss_match:
        extracted_data.append(('Valor ISS', iss_match.group(1)))
        
    inss_regex = r'INSS\n([\d,.]+)'
    inss_match = re.search(inss_regex, text)
    if inss_match:
        extracted_data.append(('Valor INSS', inss_match.group(1)))
        
    irrf_regex = r'IRRF\n([\d,.]+)'
    irrf_match = re.search(irrf_regex, text)
    if irrf_match:
        extracted_data.append(('Valor IRRF', irrf_match.group(1)))
        
    csll_regex = r'CSLL\n([\d,.]+)'
    csll_match = re.search(csll_regex, text)
    if csll_match:
        extracted_data.append(('Valor CSLL', csll_match.group(1)))
        
    cofins_regex = r'COFINS\n([\d,.]+)'
    cofins_match = re.search(cofins_regex, text)
    if cofins_match:
        extracted_data.append(('Valor COFINS', cofins_match.group(1)))
        
    pis_regex = r'PIS\n([\d,.]+)'
    pis_match = re.search(pis_regex, text)
    if pis_match:
        extracted_data.append(('Valor PIS', pis_match.group(1)))

    # Cria o DataFrame a partir dos dados extraídos
    df = pd.DataFrame(extracted_data, columns=['Campo', 'Valor'])

    # Salva os dados no arquivo Excel
    df.to_excel(output_file, encoding='utf-8', index=False)

    print(f"Arquivo '{output_file}' criado com sucesso!")

Extrair_Modelo_SP('NFS_001283_THINK DIGITAL_2.307,68.pdf', 'dados_extraidos.xlsx')
```
