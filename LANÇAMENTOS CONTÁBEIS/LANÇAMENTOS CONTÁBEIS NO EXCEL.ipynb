# DIGITE O NOME DO ARQUIVO

```python
import pandas as pd

#ESCREVA O NOME DO ARQUIVO entre os colchetes. ex: 'XP.xlsx'
df = pd.read_excel('PLANILHA.xlsx')

#veja a planilha
display(df)
```

# ATUALIZE AS CONTAS PARA O LANÇAMENTO

```python
#ex:df.loc[df['Subdivisao'] == 'Contas a pagar (C/P)' , 'DÉBITO'] = str(111111)
df.loc[df['Subdivisao'] == 'Contas a pagar (C/P)' , 'DÉBITO'] = str(23)
df.loc[df['Subdivisao'] == 'Contas a Receber (C/R)' , 'DÉBITO'] = str(2)
df.loc[df['Subdivisao'] == 'Fundo Referenciado Itaú' , 'DÉBITO'] = str(3)
df.loc[df['Subdivisao'] == 'Encargos bancários' , 'DÉBITO'] = str(4)
df.loc[df['Subdivisao'] == 'Impostos' , 'DÉBITO'] = str(5)
df.loc[df['Subdivisao'] == 'Receitas Financeiras' , 'DÉBITO'] = str(6)

df.loc[df['Subdivisao'] == 'Contas a pagar (C/P)' , 'CRÉDITO'] = str(3543)
df.loc[df['Subdivisao'] == 'Contas a Receber (C/R)' , 'CRÉDITO'] = str(2728)
df.loc[df['Subdivisao'] == 'Fundo Referenciado Itaú' , 'CRÉDITO'] = str(8639)
df.loc[df['Subdivisao'] == 'Encargos bancários' , 'CRÉDITO'] = str(1340)
df.loc[df['Subdivisao'] == 'Impostos' , 'CRÉDITO'] = str(11)
df.loc[df['Subdivisao'] == 'Receitas Financeiras' , 'CRÉDITO'] = str(2312)

#ISSO SELECIONA A COLUNA DESEJADA OU CRIA OUTRA NOVA
#E CONCATENA AS COLUNAS DESEJADAS PARA FORMAR A INFORMAÇÃO QUE DESEJA DEPOSITAR NELA
df['HISTÓRICO'] = "VLR. REF."+"  "+df['Subdivisao'] +": "+df['Comentário/descrição']
df.loc[df['Subdivisao'] == 'Contas a Receber (C/R)' , 'HISTÓRICO'] = "VLR. REF."+"  "+df['Subdivisao']+": - "+df['Nome']
df.loc[df['Subdivisao'] == 'Receitas Financeiras' , 'HISTÓRICO'] = "VLR. REF."+"  "+df['Subdivisao']+": "+df['Comentário/descrição']+" - "+df['Nome']      
df.loc[df['Subdivisao'] == 'Contas a pagar (C/P)' , 'HISTÓRICO'] = "VLR. REF."+"  "+df['Subdivisao']+": - " +df['Nome']
df.loc[df['Subdivisao'] == 'Cartão de crédito | Itaú' , 'HISTÓRICO'] = "VLR. REF."+df['Subdivisao']+": - "
```
